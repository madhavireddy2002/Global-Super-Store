# Business Questions for Exploratory Data Analysis (EDA)
### Which product category has the highest sales?


```python
df.head()
```

```python
# Total sales by category

category_sales= df.groupby('Category')['Sales'].sum()
print(category_sales)
```

```python
#Highest sales

print("Category with Highest sales:")

print(category_sales.idxmax())
```

```python
# Visualization 

category_sales.plot(kind="bar")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.title(" total sales by category")
plt.show()
```

### Observation

- Technology category generated the highest sales.
- Office Supplies contributed moderate sales.
- Furniture generated the least sales among the categories.

### Which region earns the maximum profit?

```python
#total profit by region

region_profit = df.groupby('Region')['Profit'].sum()
print(region_profit)
```

```python
# maximun profit by region

print(" Maximum profit by region:")
print(region_profit.idxmax())
```

```python
# Visualization

region_profit.plot(kind="bar")
plt.title("Maximum Profit By Region")
plt.xlabel(" Region")
plt.ylabel(" profit")

plt.show()
```

### Which shipping mode is used the most?

```python
df['Ship Mode']
```

```python
# count each shipping mode

shipping_mode=df['Ship Mode'].value_counts()
print(shipping_mode)
```

```python
# most used shipping mode 

print("Most used Shipping Mode:")
print(shipping_mode.idxmax())
```

```python
# Visualization

plt.figure(figsize=(8,5))

ax = sns.countplot(x='Ship Mode', data=df)

# Add count labels on each bar
for container in ax.containers:
    ax.bar_label(container)

plt.title("Shipping Mode Usage")
plt.xlabel("Ship Mode")
plt.ylabel("Count")

plt.show()
```

### Observation

- Standard Class is the most frequently used shipping mode.
- Same Day shipping is used the least.
- Customers prefer economical shipping methods.

### What are the yearly sales trends?

```python
df.columns
```

```python
df['Order Date'].value_counts()
```

```python
# Convert Order Date to datetime

df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
#Extract year

df['Year'] = df['Order Date'].dt.year
```

```python
yearly_sales = df.groupby('Year')['Sales'].sum()

print(yearly_sales)
```

```python
# Visualization

yearly_sales.plot(marker ='o')

plt.title("Yearly Sales Trend")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

### Top 10 products by sales?

```python
df.columns
```

```python
top_products = df.groupby("Product Name")["Sales"].sum()

top_products=top_products.sort_values(ascending =False).head(10)
print(top_products)
```

```python
# visualization

top_products.plot(kind='barh')

plt.title("Top 10 Products by Sales")

plt.xlabel("Sales")

plt.ylabel("Product Name")

plt.show()
```

### Does discount affect profit?

```python
plt.figure(figsize=(8,5))

sns.scatterplot(x="Discount", y="Profit",data=df)
plt.title(" Discount vs Profit")
plt.xlabel("Discount")
plt.ylabel("Profit")

plt.show()
```

```python
correlation = df['Discount'].corr(df['Profit'])

print("Correlation between Discount and Profit:")

print(correlation)
```

### If correlation is:

### Negative → Higher discount decreases profit.
### Positive → Higher discount increases profit.
### Near 0 → No strong relationship.

### Which customer segment contributes the most revenue?

### What is Correlation?

Correlation measures the relationship between two variables.

Its value lies between:         + 1  , -1 

```python
plt.figure(figsize=(8,6))

corr = df.corr(numeric_only=True)            # Calculates the correlation matrix of all numerical columns.

sns.heatmap(
corr,
annot=True,
cmap='coolwarm')                                     # Visualizes the correlation matrix using colors.

plt.title("Correlation Heatmap")

plt.show()
```

### Observation
The correlation heatmap was used to examine relationships among numerical variables.
Sales and Profit showed a positive correlation, indicating that higher sales generally lead to higher profits.
Discount and Profit showed a negative correlation, suggesting that higher discounts may reduce profitability.
Variables with correlation values close to zero have weak or no relationship.

### Outlier Detection

### Sales Outlier Detection

```python
plt.figure(figsize=(8,5))

sns.boxplot(x=df['Sales'])

plt.title("Outlier Detection for Sales")

plt.show()
```

### Observation
The Sales column contains several outliers.
Some orders have extremely high sales compared to most transactions.

### Profit Outlier Detection

```python
plt.figure(figsize=(8,5))

sns.boxplot(x=df['Profit'])

plt.title("Outlier Detection for Profit")

plt.show()
```

### Observation

Some transactions resulted in very high profits while others incurred losses.
